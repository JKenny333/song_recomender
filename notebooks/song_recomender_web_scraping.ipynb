## Hot 100

```python
# 1. import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 2. find url and store it in a variable
url = "https://www.billboard.com/charts/hot-100/"

response = requests.get(url)

soup = BeautifulSoup(response.content)

songs = []
artists = []
 
for song in soup.select("h3#title-of-a-story.c-title.a-no-trucate.a-font-primary-bold-s"):
    songs.append(song.get_text(strip=True))
    
for artist in soup.select("span.c-label.a-no-trucate.a-font-primary-s"):
    artists.append(artist.get_text(strip=True))

hot_100 = pd.DataFrame({"song":songs, "artist":artists})

hot_100
```

```python
hot_100.to_csv( '../data/hot_100.csv')
```

```python
alt_250_url = "https://digitaldreamdoor.com/pages/best_songs-Alternative.html"
response_alt_250 = requests.get(alt_250_url)
soup = BeautifulSoup(response_alt_250.content)
```

```python
# Find all divs with class 'list' which contain the alt_250 and artists
divs = soup.find_all('div', class_='list')

# Initialize an empty list to hold the song and artist names
alt_250 = []

# Iterate through each 'div' with class 'list'
for div in divs:
    # The children will include span, text (song title), span (artist), and br
    children = list(div.children)
    for child in children:
        if isinstance(child, NavigableString) and child.strip():
            # This is the song title text node
            song_title = child.strip()
            # The artist name is the text of the next span element, so we find the next span
            next_span = child.find_next('span')
            if next_span and next_span.string:
                artist_name = next_span.string.strip().lstrip('-').strip()
                alt_250.append({'song': song_title, 'artist': artist_name})

# Create a DataFrame from the list of dictionaries
alt_250 = pd.DataFrame(alt_250)

alt_250
```
