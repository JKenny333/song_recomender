# Final song database

```python
import pandas as pd
```

```python
#Import latest dataframe
full_af_from_csv = pd.read_csv('../data/full_af.csv')
greatest_songs_all_time_af_df = pd.read_csv('../data/greatest_songs_all_time_af.csv')
worlds_largest_af_df = pd.read_csv('../data/worlds_largest_af.csv')
r_and_b_throwbacks_af_df = pd.read_csv('../data/r_and_b_throwbacks_af.csv')
longest_playlist_ever_af_df = pd.read_csv('../data/longest_playlist_ever_af.csv')
billboard_hot_100_af_df = pd.read_csv('../data/billboard_hot_100_af.csv')
```

```python
longest_playlist_ever_af_df
```

```python
full_af = pd.concat([full_af_from_csv,longest_playlist_ever_af_df, billboard_hot_100_af_df], ignore_index=True)
full_af
```

```python
full_af.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
full_af
```

```python
full_af = full_af.drop_duplicates()

full_af.reset_index(drop=True, inplace=True)
full_af
```

```python
#export to csv
full_af.to_csv("../data/full_af.csv")
```
