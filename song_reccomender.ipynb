## Hot 100

```python
# 1. import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 2. find url and store it in a variable
url = "https://www.billboard.com/charts/hot-100/"

response = requests.get(url)

soup = BeautifulSoup(response.content)

songs = []
artists = []

#for artist in 
for song in soup.select("h3#title-of-a-story.c-title.a-no-trucate.a-font-primary-bold-s"):
    songs.append(song.get_text(strip=True))
    
for artist in soup.select("span.c-label.a-no-trucate.a-font-primary-s"):
    artists.append(artist.get_text(strip=True))

hot_100 = pd.DataFrame({"song":songs, "artist":artists})

hot_100
```

```python
Alt_250_url = "https://digitaldreamdoor.com/pages/best_songs-Alternative.html"
response_Alt_250 = requests.get(Alt_250_url)
```

```python
soup = BeautifulSoup(response_Alt_250.content)


songs = []
artists = []

#for artist in 
for song in soup.select("div.list span"):
    songs.append(song.get_text(strip=True))
    
#for artist in soup.select(""):
    #artists.append(artist.get_text(strip=True))

#hot_100 = pd.DataFrame({"song":songs, "artist":artists})
```

```python
# Find all <span> elements directly inside the div with class 'list'
for div in soup.select("div.list"):
    elements = div.contents  # This gets all the children of the div
    for i in range(0, len(elements), 6):  # Increment by 6 to jump to the next song
        # Assuming the structure is consistent with number, title, artist
        number = elements[i].get_text(strip=True) if i < len(elements) else None
        title = elements[i+2].strip() if i+2 < len(elements) else None
        artist = elements[i+4].get_text(strip=True) if i+4 < len(elements) else None

        # Append a tuple of (number, title, artist) to the songs list
        if number and title and artist:
            songs.append((number, title, artist))

# At this point, 'songs' is a list of tuples with (number, title, artist)
```
